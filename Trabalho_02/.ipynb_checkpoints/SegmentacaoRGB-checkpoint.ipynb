```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def histogram(img):
    l, c = img.shape
    histogram = np.zeros(256, dtype=np.int32)

    for i in range(l):
        for j in range(c):
            intensity = img[i, j]
            histogram[intensity] += 1
            
    return histogram
    
def show_histogram(img, img_gray):
    histogram_gray = histogram(img_gray)
    histogram_r = histogram(img[:, :, 0])
    histogram_g = histogram(img[:, :, 1])
    histogram_b = histogram(img[:, :, 2])
    
    plt.figure(figsize=(12,12))
    plt.subplot(2, 2, 1)
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.subplot(2, 2, 2)
    plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
    plt.subplot(2, 2, 3)
    plt.bar(range(256), histogram_g, color='green')
    plt.bar(range(256), histogram_b, color='blue')
    plt.bar(range(256), histogram_r, color='red')
    plt.subplot(2, 2, 4)
    plt.bar(range(256), histogram_gray)
    
def gray_image(image):
    return((0.2126 * image[:, :, 0]) + (0.7152 * image[:, :, 1]) + (0.0722 * image[:, :, 2])).astype(np.uint8)
```

```python
image = np.array(Image.open('termometro.jpg'))
(l, c, p) = image.shape

# Convertendo para grayscale
image_gray = gray_image(image)

show_histogram(image, image_gray)
```

#### Visualizando histograma de parte com o objeto de interesse (temperatura) e sem

```python
#definindo dois espaos
image_sem = image[100:300, 300:500]
image_com = image[0:100, 200:400]

image_sem_gray = gray_image(image_sem)
image_com_gray = gray_image(image_com)
```

```python
show_histogram(image_sem, image_sem_gray)
```

```python
show_histogram(image_com, image_com_gray)
```

#### Canal vermelho  muito presente na regio de interesse
Limiar ser baseado no canal vermelho, sendo assim threshold > 250

A luz branca abaixo do valor pode possuir o canal vermelho acima do limiar, essa parte provavelmente tambm possui os canais G e B acima do limiar, ento  necessrio tratar essa luz

```python
def red_threshold(image, threshold):
    
    l, c, p = image.shape
    red_bin = np.zeros(shape=(l, c), dtype=np.uint8)
    
    for i in range (l):
        for j in range (c):
            
            value_r = image[i,j,0]
            value_g = image[i,j,1]
            value_b = image[i,j,2]
            
            if value_r >= threshold and value_g < threshold and value_b < threshold:
                red_bin[i,j] = 1
            else:
                red_bin[i,j] = 0
                
    return red_bin
            
```

```python
threshold = 250
bin_image = red_threshold(image, threshold)
not_bin = np.invert(bin_image)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(1, 2, 1)
plt.imshow(bin_image, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(not_bin, cmap='gray')
```
